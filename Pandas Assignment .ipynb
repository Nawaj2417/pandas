```python
Ans:- The difference between the functions loc and iloc in pandas are:-
        'loc' uses label-based indexing and 'iloc uses integer-based indexing'
example:
```

```python
df = pd.DataFrame({'A': [1,2,3], 'B': [5,6,7]},index= ['a','b','c'])
df.loc[['a','b'], 'A']
```

```python
df.iloc[:2,1]
```

```python

df1 = pd.DataFrame({'A': [1,2,3,4], 'B': [5,6,7,5]}, index = [0,1,2,3])
```

```python
new_df.iloc[2]
```

```python
From the above code , we see that 'new_df.loc[2]' returns the row with label '2' and 'new_df.iloc[2]' returns
the row with integer index '2'
```

```python
# Q6. What do you understand about the windows function in pandas and list the types of windows
# functions?
```

```python
Ans:- In pandas, a window function is also known as rolling or moving function is a way to apply a 
    function to a rolling window of data in a DataFrame or Series. It takes a window of data, performs 
    a computation and then moves the forward by one unit and repeats the computation, generating a new set of results.
    
    The types of windows function are:- 
        1. Rolling functions
        2. Expanding Functions
        3. Exponential weighted function
        4. Rolling and expanding apply function
        5. Rolling and expanding aggregation functions.
```

```python
# Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
# calculates the difference between them in days, hours, and minutes using Pandas time delta. The
# program should prompt the user to enter the dates and display the result.
```

```python
import pandas as pd
a = input("Enter the first date (YYYY-MM-DD)")
b = input("Enter the second date (YYYY-MM-DD)")

a = pd.to_datetime(a)
b = pd.to_datetime(b)

time_diff = b -a 

days = time_diff.days
hours = time_diff.seconds //3600
minutes = (time_diff.seconds //60) % 60

new_date_time = days,hours, minutes
```

```python
new_date_time
```
